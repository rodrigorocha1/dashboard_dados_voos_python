# Funções

- Conversão de atributos

```python
def converter_atributos(dataframe: pd.DataFrame) -> pd.DataFrame:
   
    dataframe['SIGLA_ICAO_EMPRESA_AEREA'] = dataframe['SIGLA_ICAO_EMPRESA_AEREA'].astype('string')
    dataframe['PARTIDA_PREVISTA'] = pd.to_datetime(dataframe['PARTIDA_PREVISTA'],format='%Y-%m-%d %H:%M:%S')
    dataframe['PARTIDA_REAL'] = pd.to_datetime(dataframe['PARTIDA_REAL'], format='%Y-%m-%d %H:%M:%S')
    dataframe['CHEGADA_PREVISTA'] = pd.to_datetime(dataframe['CHEGADA_PREVISTA'], format='%Y-%m-%d %H:%M:%S')
    dataframe['CHEGADA_REAL'] = pd.to_datetime(dataframe['CHEGADA_REAL'], format='%Y-%m-%d %H:%M:%S')
    dataframe[['NUMERO_VOO', 'CODIGO_TIPO_LINHA', 'SIGLA_ICAO_AEROPORTO_ORIGEM', 'SIGLA_ICAO_AEROPORTO_DESTINO', 'SITUACAO_VOO']] = dataframe[['NUMERO_VOO', 'CODIGO_TIPO_LINHA', 'SIGLA_ICAO_AEROPORTO_ORIGEM', 'SIGLA_ICAO_AEROPORTO_DESTINO', 'SITUACAO_VOO']].astype('string')
    return dataframe
```

- Obter faixa de atraso

```python
def obter_faixa_atraso(minutos: int) -> str:
    if minutos <= 0:
        faixa = 'No horário'
    elif  0 > minutos < 15:
        faixa = 'Atraso de até 15 minutos'
    elif 15 >= minutos < 30:
        faixa = 'Atraso de até 30 minutos'
    elif 30 >= minutos < 60:
        faixa = 'Atraso de até 1 hora'
    elif 60 >= minutos < -120:
        faixa = 'Atraso de até 2 horas'
    elif 120 >= minutos < -240:
        faixa = 'Atraso de até 4 horas'
    else:
        faixa = 'Atraso de mais de 4 horas'
    return faixa
```

- Obter Tradução

```python
def obter_traducao(dia: str):
    traducao_dias = {
        'Monday': 'Segunda-feira',
        'Tuesday': 'Terça-feira',
        'Wednesday': 'Quarta-feira',
        'Thursday': 'Quinta-feira',
        'Friday': 'Sexta-feira',
        'Saturday': 'Sábado',
        'Sunday': 'Domingo'
    }
    return traducao_dias.get(dia)

```

- Obter Fim de Semana

```python
def obter_fim_de_semana(indice: int) -> bool:
    if indice in [6, 5]:
        flag = True
    else:
        flag = False
    return flag
```
