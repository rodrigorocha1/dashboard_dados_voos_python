# FunÃ§Ãµes

- ConversÃ£o de atributos

```python
def converter_atributos(dataframe: pd.DataFrame) -> pd.DataFrame:
   
    dataframe['SIGLA_ICAO_EMPRESA_AEREA'] = dataframe['SIGLA_ICAO_EMPRESA_AEREA'].astype('string')
    dataframe['PARTIDA_PREVISTA'] = pd.to_datetime(dataframe['PARTIDA_PREVISTA'],format='%Y-%m-%d %H:%M:%S')
    dataframe['PARTIDA_REAL'] = pd.to_datetime(dataframe['PARTIDA_REAL'], format='%Y-%m-%d %H:%M:%S')
    dataframe['CHEGADA_PREVISTA'] = pd.to_datetime(dataframe['CHEGADA_PREVISTA'], format='%Y-%m-%d %H:%M:%S')
    dataframe['CHEGADA_REAL'] = pd.to_datetime(dataframe['CHEGADA_REAL'], format='%Y-%m-%d %H:%M:%S')
    dataframe[['NUMERO_VOO', 'CODIGO_TIPO_LINHA', 'SIGLA_ICAO_AEROPORTO_ORIGEM', 'SIGLA_ICAO_AEROPORTO_DESTINO', 'SITUACAO_VOO']] = dataframe[['NUMERO_VOO', 'CODIGO_TIPO_LINHA', 'SIGLA_ICAO_AEROPORTO_ORIGEM', 'SIGLA_ICAO_AEROPORTO_DESTINO', 'SITUACAO_VOO']].astype('string')
    return dataframe
```

- Obter faixa de atraso

```python
def obter_faixa_atraso(minutos: int) -> str:
    if minutos <= 0:
        faixa = 'No horÃ¡rio'
    elif  0 > minutos < 15:
        faixa = 'Atraso de atÃ© 15 minutos'
    elif 15 >= minutos < 30:
        faixa = 'Atraso de atÃ© 30 minutos'
    elif 30 >= minutos < 60:
        faixa = 'Atraso de atÃ© 1 hora'
    elif 60 >= minutos < -120:
        faixa = 'Atraso de atÃ© 2 horas'
    elif 120 >= minutos < -240:
        faixa = 'Atraso de atÃ© 4 horas'
    else:
        faixa = 'Atraso de mais de 4 horas'
    return faixa
```

- Obter TraduÃ§Ã£o

```python
def obter_traducao(dia: str):
    traducao_dias = {
        'Monday': 'Segunda-feira',
        'Tuesday': 'TerÃ§a-feira',
        'Wednesday': 'Quarta-feira',
        'Thursday': 'Quinta-feira',
        'Friday': 'Sexta-feira',
        'Saturday': 'SÃ¡bado',
        'Sunday': 'Domingo'
    }
    return traducao_dias.get(dia)

```

- Obter Fim de Semana

```python
def obter_fim_de_semana(indice: int) -> bool:
    if indice in [6, 5]:
        flag = True
    else:
        flag = False
    return flag
```

```python
base_voo = converter_atributos(base_voo)
```

```python
base_voo['DIA_DA_SEMANA_PARTIDA_PREVISTA'] = base_voo['PARTIDA_PREVISTA'].dt.day_name()
base_voo['INDICE_SEMANA_PARTIDA_PREVISTA'] = base_voo['PARTIDA_PREVISTA'].dt.dayofweek
base_voo['DIA_DA_SEMANA_PARTIDA_REAL'] = base_voo['PARTIDA_REAL'].dt.day_name()
base_voo['INDICE_SEMANA_PARTIDA_REAL'] = base_voo['PARTIDA_REAL'].dt.dayofweek
base_voo['DIA_DA_SEMANA_CHEGADA_REAL'] = base_voo['CHEGADA_REAL'].dt.day_name()
base_voo['DIA_DA_SEMANA_CHEGADA_PREVISTA'] = base_voo['CHEGADA_PREVISTA'].dt.day_name()
```

```python
base_voo['DIA_DA_SEMANA_PARTIDA_PREVISTA'] = base_voo['DIA_DA_SEMANA_PARTIDA_PREVISTA'].apply(obter_traducao)
base_voo['DIA_DA_SEMANA_PARTIDA_REAL'] = base_voo['DIA_DA_SEMANA_PARTIDA_REAL'].apply(obter_traducao)
base_voo['DIA_DA_SEMANA_CHEGADA_REAL'] = base_voo['DIA_DA_SEMANA_CHEGADA_REAL'].apply(obter_traducao)
base_voo['DIA_DA_SEMANA_CHEGADA_PREVISTA'] = base_voo['DIA_DA_SEMANA_CHEGADA_PREVISTA'].apply(obter_traducao)
```

```python
base_voo['INDICE_SEMANA_PARTIDA_REAL'] = base_voo['PARTIDA_REAL'].dt.dayofweek
base_voo['INDICE_SEMANA_PARTIDA_PREVISTA'] = base_voo['PARTIDA_PREVISTA'].dt.dayofweek
base_voo['INDICE_SEMANA_CHEGADA_REAL'] = base_voo['CHEGADA_REAL'].dt.dayofweek
base_voo['INDICE_SEMANA_CHEGADA_PREVISTA'] = base_voo['CHEGADA_PREVISTA'].dt.dayofweek
```

```python
base_voo['FIM_DE_SEMANA_CHEGADA_REAL'] = base_voo['INDICE_SEMANA_CHEGADA_REAL'].apply(obter_fim_de_semana)
base_voo['FIM_DE_SEMANA_PARTIDA_REAL'] = base_voo['INDICE_SEMANA_CHEGADA_REAL'].apply(obter_fim_de_semana)
```
