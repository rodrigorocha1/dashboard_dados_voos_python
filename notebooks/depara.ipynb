```python
import pandas as pd
import pickle
```

```python
def salvar_picke(lista_arquivo, nome_pkl):
    with open(f'../src/depara/{nome_pkl}.pkl', 'wb') as arq:
        pickle.dump(lista_arquivo, arq)
```

```python
base_depara = pd.read_parquet('../data/processed/df_processed_v2.parquet')
```

```python
base_depara.columns
```

```python
base_depara['NOME_COMPLETO_EMPRESA'] = base_depara['SIGLA_ICAO_EMPRESA_AEREA'] + '-'   + base_depara['NOME_EMPRESA']
base_depara
```

```python
lista_companias = base_depara['NOME_COMPLETO_EMPRESA'].to_list()
```

```python
lista_companias = list(set(lista_companias))
lista_companias.sort(reverse=False)
```

```python
salvar_picke(lista_arquivo=lista_companias, nome_pkl='lista_companias')
```
