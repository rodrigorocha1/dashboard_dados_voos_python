```python
columns = ['Sigla ICAO Empresa Aérea', 'Número Voo', 
       'Código Tipo Linha', 'Sigla ICAO Aeroporto Origem',
       'Sigla ICAO Aeroporto Destino', 'Partida Prevista', 'Partida Real',
       'Chegada Prevista', 'Chegada Real', 'Situação Voo']
```

```python
def converter_atributos(dataframe: pd.DataFrame) -> pd.DataFrame:
   
    dataframe['Sigla ICAO Empresa Aérea'] = dataframe['Sigla ICAO Empresa Aérea'].astype('string')
    dataframe['Partida Prevista'] = dataframe['Partida Prevista'].astype('string')
    dataframe['Partida Prevista'] = pd.to_datetime(dataframe['Partida Prevista'],format='%d/%m/%Y %H:%M')
    dataframe['Partida Real'] = pd.to_datetime(dataframe['Partida Real'], format='%d/%m/%Y %H:%M')
    dataframe['Chegada Prevista'] = pd.to_datetime(dataframe['Chegada Prevista'], format='%d/%m/%Y %H:%M')
    dataframe['Chegada Real'] = pd.to_datetime(dataframe['Chegada Real'], format='%d/%m/%Y %H:%M')
    dataframe[['Número Voo', 'Código Tipo Linha', 'Sigla ICAO Aeroporto Origem', 'Sigla ICAO Aeroporto Destino', 'Situação Voo']] = dataframe[['Número Voo', 'Código Tipo Linha', 'Sigla ICAO Aeroporto Origem', 'Sigla ICAO Aeroporto Destino', 'Situação Voo']].astype('string')
    return dataframe
```

```python
def abrir_dataframe(caminho: str, encoding: str ) -> pd.DataFrame:
    dataframe = pd.read_csv(caminho, sep=';', encoding='utf-8', usecols=columns)
    return dataframe
```

```python
colunas_data = ['PARTIDA_PREVISTA', 'PARTIDA_REAL', 'CHEGADA_PREVISTA', 'CHEGADA_REAL']
```

```python
for coluna_data in colunas_data:
    result[coluna_data].fillna('1900-01-01', inplace=True)
    result[f'MES_{coluna_data}'] = pd.to_datetime(result[coluna_data]).dt.month
    result[f'NOME_MES_{coluna_data}'] = pd.to_datetime(result[coluna_data]).dt.strftime('%B')
```

```python
result['MINUTOS_ATRASADO_PARTIDA'] = ( pd.to_datetime(result['PARTIDA_REAL']) -  pd.to_datetime(result['PARTIDA_PREVISTA']) ).dt.total_seconds() / 60
result['MINUTOS_ATRASADO_CHEGADA'] = (pd.to_datetime(result['CHEGADA_REAL']) - pd.to_datetime(result['CHEGADA_PREVISTA']) ).dt.total_seconds() / 60
```
